```python
import pandas as pd
df=pd.read_csv('laliga.csv')
df=df.replace('-',0)
df
```

```python
df['DebutNew']=df['Debut'].str[:4] #making a new column for debut in order to slice values based on year
df[(df['DebutNew']>='1930') & (df['DebutNew']<='1980')]
```

```python
df['Points']=df['Points'].astype(int)
df.sort_values('Points',ascending=False,inplace=True)
df[['Team','Points']].head()
```

```python
def Goal_diff_count():
    df['GoalDiff']=df['GoalsFor'].astype(int)-df['GoalsAgainst'].astype('int')
    print(df[['Team','GoalDiff']])
    print('Max Goal Difference: ',max(df['GoalDiff']))
    print('Min Goal Difference: ',min(df['GoalDiff']))
Goal_diff_count()
```

```python
df['WinningPercent']=df['GamesWon'].astype(int)/df['GamesPlayed'].astype(int)*100
df['WinningPercent']=df['WinningPercent'].fillna(0)
df.sort_values('WinningPercent',ascending=False,inplace=True)
df[['Team','WinningPercent']].head()
```
